* registered : nombre de locations par  des utilisateurs enregistrés/jour
* count : nombre de locations = casual + registered

## Analyse corrélation

```python
df['year'] = pd.to_datetime(df['datetime']).dt.year
df['month'] = pd.to_datetime(df['datetime']).dt.month
df['day'] = pd.to_datetime(df['datetime']).dt.day
df['hour'] = pd.to_datetime(df['datetime']).dt.hour
df['dayofweek'] = pd.to_datetime(df['datetime']).dt.weekday
```

```python
df['date_data'] = df['year'].map(str)+ '-' +df['month'].map(str)
df['date_data'] = pd.to_datetime(df['date_data'], format='%Y-%m').dt.strftime('%Y-%m')

df_group_month_mean = df.groupby('date_data',as_index=False).mean()
df_group_month_count = df.groupby('date_data',as_index=False).count()
df_group_month_sum = df.groupby('date_data',as_index=False).sum()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
